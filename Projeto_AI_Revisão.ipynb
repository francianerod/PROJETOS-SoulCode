```python
# Para ver todas as categorias dos atributos
categorico_val = []
continuo_val = []

# Verificação de dados únicos em cada coluna e separação em dados categóricos e continuos
for column in df.columns:
    print('==============================')
    print(f"{column} : {df[column].unique()}")

    if len(df[column].unique()) <= 10:
        categorico_val.append(column)
    else:
        continuo_val.append(column)
```

```python
# Função Regressão adaptada para o SGDRegressor

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def print_regression_score(model, X_train, y_train, X_test, y_test, train=True):
    if train:
        y_pred_train = model.predict(X_train)
        mse_train = mean_squared_error(y_train, y_pred_train)
        r2_train = r2_score(y_train, y_pred_train)
        mae_train = mean_absolute_error(y_train, y_pred_train)
        rmse_train = mean_squared_error(y_train, y_pred_train, squared=False)  # Calcula o RMSE
        # Calcula o MAPE
        mape_train = np.mean(np.abs((y_train - y_pred_train) / y_train)) * 100

        print("Resultado do treinamento:\n============================")
        print(f"Mean Absolute Error (MAE) - Treinamento: {mae_train:.2f}")
        print(f"Mean Squared Error (MSE) - Treinamento: {mse_train:.2f}")
        print(f"Root Mean Squared Error (RMSE) - Treinamento: {rmse_train:.2f}")
        print(f"Mean Absolute Percentage Error (MAPE) - Treinamento: {mape_train:.2f}%")
        print(f"R-squared (R2) - Treinamento: {r2_train:.2f}")

    else:
        y_pred_test = model.predict(X_test)
        mse_test = mean_squared_error(y_test, y_pred_test)
        r2_test = r2_score(y_test, y_pred_test)
        mae_test = mean_absolute_error(y_test, y_pred_test)
        rmse_test = mean_squared_error(y_test, y_pred_test, squared=False)  # Calcula o RMSE

        # Calcula o MAPE
        mape_test = np.mean(np.abs((y_test - y_pred_test) / y_test)) * 100

        print("Resultado do teste:\n============================")
        print(f"Mean Absolute Error (MAE) - Teste: {mae_test:.2f}")
        print(f"Mean Squared Error (MSE) - Teste: {mse_test:.2f}")
        print(f"Root Mean Squared Error (RMSE) - Teste: {rmse_test:.2f}")
        print(f"Mean Absolute Percentage Error (MAPE) - Teste: {mape_test:.2f}%")
        print(f"R-squared (R2) - Teste: {r2_test:.2f}")
```

```python
# Avaliando o modelo SGDRegressor
print_regression_score(regressor, X_train, y_train, X_test, y_test, train=True)
print_regression_score(regressor, X_train, y_train, X_test, y_test, train=False)
```

```python
# r-quadrado ajustado
def adjusted_r2(y_test, y_pred,X_train):
    from sklearn.metrics import r2_score
    adj_r2 = (1 - ((1 - r2_score(y_test, y_pred)) * (len(y_test) - 1)) / (len(y_test) - X_train.shape[1] - 1))
    return adj_r2
```

```python
# Imprimindo r-quadrado ajustado
print(round(adjusted_r2(y_test,y_pred,X_train),2))
```

```python
# Importando Bibliotecas
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
```

```python
# Separação da base em Treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=42)
```

Os parâmetros do modelo XGBClassifier podem ser ajustados para melhorar o desempenho do modelo. Os parâmetros mais importantes a serem ajustados são:

*  n_estimators: O número de árvores de decisão no modelo. Um valor maior de n_estimators geralmente resulta em um modelo mais preciso, mas também pode levar a um aumento do tempo de treinamento.

* learning_rate: A taxa de aprendizado do modelo. Um valor menor de learning_rate geralmente resulta em um modelo mais estável, mas também pode levar a um aumento do tempo de treinamento.

*  max_depth: A profundidade máxima das árvores de decisão no modelo. Um valor maior de max_depth geralmente resulta em um modelo mais preciso, mas também pode levar a um aumento do risco de overfitting.

```python
# Criando o modelo
clf = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=3)

# Treinando o modelo
clf.fit(X_train, y_train)

# Fazendo previsões
y_pred = clf.predict(X_test)
```

```python
# Função Classificação
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def print_score(clf, X_train, y_train, X_test, y_test, train=True):
    if train:
        pred = clf.predict(X_train)
        clf_report = pd.DataFrame(classification_report(y_train, pred, output_dict=True))
        print("Resultado do treinamento:\n================================================")
        print(f"Accuracy Score: {accuracy_score(y_train, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"Relatório de classificação:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_train, pred)}\n")

    elif train==False:
        pred = clf.predict(X_test)
        clf_report = pd.DataFrame(classification_report(y_test, pred, output_dict=True))
        print("Resultado do teste:\n================================================")
        print(f"Accuracy Score: {accuracy_score(y_test, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"Relatório de classificação:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_test, pred)}\n")
```

```python
# Relatório de resultados
print_score(clf, X_train, y_train, X_test, y_test, train=True)
print_score(clf, X_train, y_train, X_test, y_test, train=False)
```
