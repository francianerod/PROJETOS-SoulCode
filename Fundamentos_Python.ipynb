## **Estrutura de Condição: IF, ELIF, ELSE**

Instrução "if":
*  A instrução "if" é usada para executar um bloco de código se uma determinada condição for verdadeira. A sintaxe geral é a seguinte:

```python
'''
if condição:
    # bloco de código a ser executado se a condição for verdadeira
'''
```

```python
idade = 19
if idade >= 18:
    print("Você é maior de idade.")
```

```python
idade = 18
if idade >= 18:
    print("Você é maior de idade.")
```

```python
idade = 17
if idade >= 18:
    print("Você é maior de idade.")
```

Instrução "else":
*  A instrução "else" é usada em conjunto com a instrução "if" e especifica um bloco de código a ser executado se a condição do "if" não for verdadeira. A sintaxe geral é a seguinte:

```python
'''
if condição:
    # bloco de código a ser executado se a condição for verdadeira
else:
    # bloco de código a ser executado se a condição não for verdadeira
'''
```

```python
idade = int(input('Digite sua idade: '))

if idade >= 18:
    print("Você é maior de idade.")
else:
    print("Você é menor de idade.")
```

Instrução "elif":
*  A instrução "elif" é uma abreviação de "else if" e permite testar múltiplas condições em sequência. Ela é usada quando há mais de duas possibilidades a serem consideradas. A sintaxe geral é a seguinte:

```python
'''
if condição1:
    # bloco de código a ser executado se a condição1 for verdadeira
elif condição2:
    # bloco de código a ser executado se a condição2 for verdadeira
else:
    # bloco de código a ser executado se nenhuma das condições anteriores for verdadeira
'''
```

```python
nota = float(input('Digite sua nota: '))

if nota >= 90:
    print("Sua nota é A.")

elif nota >= 80:
    print("Sua nota é B.")

elif nota >= 70:
    print("Sua nota é C.")

else:
    print("Sua nota é D.")
```

Exemplo:
*  Escreva um programa que verifique se um número digitado pelo usuário é positivo, negativo ou zero.

Passo a passo:

1. Solicite ao usuário que digite um número.
2. Armazene o número digitado em uma variável chamada "numero".
3. Verifique se o número digitado é igual a zero. Se sim, exiba a mensagem "O número digitado é zero."
4. Caso contrário, verifique se o número é maior que zero. Se sim, exiba a mensagem "O número digitado é positivo."

Sequências: no Python sequências são conjuntos de elementos ordenados que podem ser acessados em sua ordem (o que chamamos de iteráveis).

No módulo básico do Python existem quatro tipos de dados em coleções:

*  Lista (list): ordenada e mutável, pode ter membros duplicados.

Dicionários (dictionaries) são coleções de dados armazenados em pares chave: valor (key: value). A coleção é mutável, ordenada e não admite valores duplicados. A chave de um dicionário funciona como um índice que permite a recuperação do valor a ele associado. Eles têm a forma geral de dict = {chave_1:valor_1, …, chave_n:valor_n}

```python
# Declarar um dicionário
pessoa = {'nome': 'Joaquim da Silva',
          'telefone': '(11)33333-3333',
          'cpf': '000.000.000-00'
          }

# Coloca o nome da chave
print(pessoa['nome'])
print(pessoa['telefone'])
print(pessoa['cpf'])
```

```python
# Exemplo 2: Lista dentro do dicionário
telefones = ['(11)9.0000-0000', '(21)9.0000-0000', '(31)9.0000-0000']

telefones = {'sp':['(11)9.0000-0000', '(11)9.0000-0001','(11)9.0000-0002'],
             'rj':'(21)9.0000-0000',
             'mg': '(31)9.0000-0000' }

# Informação pela chave
print(telefones['sp'])
```

## **Estrutura de Repetição: WHILE**

A estrutura while é uma das formas mais comuns de controle de fluxo em programação, permitindo que um bloco de código seja executado repetidamente enquanto uma condição especificada seja verdadeira. É crucial garantir que a condição eventualmente se torne falsa, pois, se não o fizer, seu programa ficará preso em um "loop infinito", consumindo recursos da máquina indefinidamente.

```python
'''
while condição:
    blocodecodigo1
'''
```

O while não possui um contador interno. Em vez disso, ele depende de uma condição para continuar ou interromper a repetição. Isso o torna muito flexível e útil.

```python
'''
i = 1, 0 , -1 ou outros (inicio da contagem, também chamada de váriável de indexação ou controle)

while condição (normalmente um operador relacional com a condição de parada):
    bloco de codigo enquanto a condição é verdadeira
    operadores de atribuição
'''
```

```python
# Exemplo 1: Contagem Regressiva de 5 a 1:
i = 5                   # Comece a partir desse valor (váriável de controle)
while i >= 1:           # Enquanto i for maior ou igual a 1 (operador relacional com a condição de parada)
    print(i)            # imprima i (ou bloco de código)
    i -= 1              # De forma que i = i-1 (Incremento necessário para parar a iteração)
```

```python
# Exemplo 2: Soma dos números de 1 a 10:
soma = 0
numero = 1

while numero <= 10:
    soma += numero
    print(soma)
    numero += 1
```

```python
# Exemplo 3: Loop com entrada do usuário:

senha_correta = "1234"
senha = input("Digite a senha: ")

while senha != senha_correta:
    senha = input("Senha incorreta. Tente novamente: ")

print("Senha correta! Acesso concedido.")
```

Em programação, você tem o controle total dentro de um loop while. Às vezes, você pode desejar sair prematuramente de um loop ou pular uma iteração e passar para a próxima. Para isso, existem duas palavras-chave importantes: break e continue

Neste exemplo, o loop while começa em 1 e imprime os números de 1 a 4. Quando o valor da variável numero se torna igual a 5, a condição if numero == 5 é verdadeira, e o break é acionado, fazendo com que o loop seja encerrado imediatamente.

```python
# Exemplo de break:
# Vamos usar um loop while para imprimir números de 1 a 10, mas queremos sair do loop assim que chegarmos a 5.

numero = 1
while numero <= 10:
    if numero == 5:
        break
    print(numero)
    numero += 1
```

Neste exemplo, o loop while começa em 1 e imprime os números ímpares de 1 a 9. Quando o valor da variável numero é par (ou seja, numero % 2 == 0), o continue é acionado. Isso faz com que o loop pule a impressão do número par e continue com a próxima iteração, imprimindo apenas números ímpares.

```python
# Exemplo de continue com while:
# Vamos criar um loop while que imprime todos os números ímpares de 1 a 10, pulando os números pares.

numero = 1
while numero <= 10:
    if numero % 2 == 0:
        numero += 1
        continue
    print(numero)
    numero += 1
```

Recomendações de uso: Use o while quando a condição de término não for conhecida antecipadamente e quando estiver esperando eventos ou interações do usuário para realizar validações.

## **Estrutura de Repetição: FOR**

O loop for é uma estrutura de repetição que permite executar um bloco de código várias vezes. Ele é amplamente usado para percorrer elementos em sequências, como listas, tuplas e strings. O loop for é uma escolha natural quando você sabe quantas vezes deseja repetir uma ação.

```python
'''
for variável in sequência:
    bloco de codigo a ser repetido
'''
```

*  sequência: podem ser listas, tuplas e strings

*  A função range poderá ter até 3 argumentos
for i in range (inicio_do_contador, condição_de_parada, incremento ou decremento)

*  Exemplo comparativo: Imprima na tela os números de 0 a 9 com while e for

```python
# Usando While
i = 0
while i < 10:
    print(i)
    i += 1
```

```python
# 1° Forma - utilizando o range com apenas um argumento - condição de parada com incremento 1 por padrão
for i in range(10):
  print(i)
```

```python
# 2° Forma - utilizando o range com apenas 2 argumentos - inicio da contagem e a condição de parada
for i in range(1, 11):
  print(i)
```

```python
# 3° Forma - utilizando o range com apenas 3 argumentos - inicio da contagem, condição de parada e incremento
for i in range(5, 20, 3):
  print(i)
```

```python
# 4° Forma: Calculando a soma dos números de 1 a 10:
soma = 0
for i in range(1, 11):
    soma += i
print("A soma dos números de 1 a 10 é:", soma)
```

```python
# Exemplo de iteração de string
palavra = "Python"

for i in palavra:
    print(i)
```

```python
# Exemplo de iteração de tupla
tupla = (10, 20, 30, 40, 50)
for i in tupla:
    print(i)
```

```python
# Exemplo de iteração de listas
nomes = ["Alice", "Bob", "Charlie"]

for nome in nomes:
    print(nome)
```

```python
# Exemplo de iteração de dicionários
frutas = {"maçã": 3,
          "banana": 5,
          "laranja": 2
         }

for chave, valor in frutas.items():
    print(f'{chave}: {valor}')
```

```python
# 6. Função com cálculos, parâmetros, return e comentários
def formula_epg(m,h,g=10):
    '''
    Calcula energia potencial gravitacional
    argumentos ou parametros
    m = massa, entrada como uma variável float
    h = altura, entrada como uma variável float

    Argumento opcional
    g: aceleração gravitacional com valor padrão 10
    '''
    e = m*g*h
    return e
```
