## **Problemas de Classificação**

```python
# Instalação de pacote
!pip install lazypredict
```

```python
# Abertura de bibliotecas
import pandas as pd
import lazypredict
from lazypredict.Supervised import LazyClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
# Ignorar avisos
import warnings
warnings.filterwarnings("ignore")
```

```python
# Configuração da quantidade de colunas para aparecer em um Datraframe
pd.set_option('display.max_columns',100)
```

```python
# Função geral
def sklearn_to_df(sklearn_dataset):
    df = pd.DataFrame(sklearn_dataset.data, columns=sklearn_dataset.feature_names)
    df['target'] = pd.Series(sklearn_dataset.target)
    return df
```

Informações da base de dados: https://archive.ics.uci.edu/dataset/17/breast+cancer+wisconsin+diagnostic

```python
# Aplicação da função
df_saude = sklearn_to_df(load_breast_cancer())
df_saude.head()
```

```python
# Separação da base em Treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(df_saude.drop('target', axis=1),
                                                    df_saude['target'],
                                                    test_size=0.2)
```

```python
# Criando um objeto do pacote Lazy
clf = LazyClassifier(verbose=0,
                     ignore_warnings=True,
                     custom_metric=None)
```

```python
# Ajustando os dados de treinamento e teste
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
models
```

```python
# Abertura de bibliotecas
import pandas as pd
import lazypredict
from lazypredict.Supervised import LazyRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
```
